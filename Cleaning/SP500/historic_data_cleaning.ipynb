```python
import pandas as pd
import numpy as np
from datetime import datetime
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.read_csv('historic_candle_stick.csv')
```

```python
df
```

```python
df['date']=df['date'].apply(lambda row: datetime.strptime(row, "%m/%d/%y"))
df
```

```python
# drop unneeded columns
df.drop(['open','high','low','close','candlestick_match_count'], axis=1, inplace=True)
df
```

```python
df['candlestick_pattern'].unique()
```

```python
y = pd.get_dummies(df.candlestick_pattern)
y
```

```python
y.insert(loc = 0, column = 'date', value = df['date'])
y
```

```python
# save file as csv
y.to_csv('candlestick_patterns.csv')
```
